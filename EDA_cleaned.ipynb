```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("FINALL_cleaned_student_stress.xlsx")
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
sns.countplot(x='how_would_you_rate_your_current_stress_level?', data=df)
plt.title("Stress Level Distribution (0=Low,1=Moderate,2=High)")
plt.show()
```

```python
plt.hist(df['age'], bins=10)
plt.title("Age Distribution")
plt.show()
```

```python
plt.hist(df['average_sleep_hours_per_night'], bins=10)
plt.title("Sleep Hours Distribution")
plt.show()
```

```python
plt.hist(df['average_study_hours_per_day'], bins=10)
plt.title("Study Hours Distribution")
plt.show()
```

```python
sns.boxplot(x='how_would_you_rate_your_current_stress_level?', y='average_sleep_hours_per_night', data=df)
plt.title("Stress vs Sleep Hours")
plt.show()
```

```python
sns.boxplot(x='how_would_you_rate_your_current_stress_level?', y='average_study_hours_per_day', data=df)
plt.title("Stress vs Study Hours")
plt.show()
```

```python
sns.boxplot(x='how_would_you_rate_your_current_stress_level?', y='academic_pressure_level', data=df)
plt.title("Stress vs Academic Pressure")
plt.show()
```

```python
sns.boxplot(x='how_would_you_rate_your_current_stress_level?', y='daily_social_media_usage', data=df)
plt.title("Stress vs Social Media Usage")
plt.show()
```

```python
sns.boxplot(x='how_would_you_rate_your_current_stress_level?', y='physical_activity_per_week', data=df)
plt.title("Stress vs Physical Activity")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
